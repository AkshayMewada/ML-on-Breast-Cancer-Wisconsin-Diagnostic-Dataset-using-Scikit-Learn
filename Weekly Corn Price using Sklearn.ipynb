```python
# Importing Modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
import seaborn as sns
```

```python
# Reading Data from csv
data = pd.read_csv('E:\Akshay\python\Machine\exdata\Corn\corn_OHLC2013-2017.txt', sep=",", header=None)
data.columns =['Date', 'Open Price', 'High Price', 'Low Price', 'Close Price']
data
```

```python
# Describing Data
data.describe()
```

```python
# Checking Null Values
data.isnull().sum()
```

```python
# Creating New column showing week number
data['week'] = [i for i in range(1,len(data['Date'])+1)]
data.tail()
```

```python
# Checking Correlation of Data
data.corr()
```

```python
# plotting Prices using Seaborn FaceGrid
sns.factorplot(data=data[['Open Price','High Price','Low Price','Close Price']])
```

```python
# Train Data
week = data['week'].values.reshape(len(data['week']),1)

# Target Data
Open_Price = data['Open Price']
High_Price = data['High Price']
Low_Price = data['Low Price']
Close_Price = data['Close Price'] 
```

```python
# Creating model for Prediction
from sklearn.svm import SVR

def SVResgressor(train,target):
    # using SVR model for regression
    
    # Default RBF kernerl
    svr = SVR(C = 1e3,gamma = 0.01)
    svr.fit(train,target)
    pred = svr.predict(train)
    
    # plotting prediction and real price
    plt.figure(figsize=(15,5))
    plt.scatter(train,target,label ='Real Price')
    plt.plot(train,pred,'r',label ='Predicted Pricee')
    plt.legend()
```

```python
# Open price of corn per week
SVResgressor(week,Open_Price)
```

```python
# High price of corn per week
SVResgressor(week,High_Price)
```

```python
# Low price of corn per week
SVResgressor(week,Low_Price)
```

```python
# Close price of corn per week
SVResgressor(week,Close_Price)
```
